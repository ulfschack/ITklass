Läs om **elif** på egen hand (t.ex här https://docs.python.org/3/tutorial/controlflow.html), och skriv sedan kod som motsvarar följande diagram<br><br>
Obs, ni får gärna ha variabler som *name* och *age* som tilldelas sitt värde, direkt i koden, så ni slipper hålla på med **input** hela tiden.* T.ex<br>name = 'Ulf"<br>age = 3000

```python
from IPython.display import Image
Image(filename="flow3.jpg", width = 400)
```

```python
# input-commandot (data från användaren) är bökigt och ska översättas från ...
# ... string till int etc. Därför är det ok att skriva variabelvärdena direkt i koden
name = 'Carol'
age = 53
if name == 'Alice':
    print('Hi, Alice.')
elif age < 12:
    print('You are not Alice, kiddo.')
elif age > 2000:
    print('Unlike you, Alice is not an undead, immortal vampire.')
elif age > 100:
    print('You are not Alice, grannie.')
    
# Den uppmärksamme ser att inget skrivs ut för åldrarna age = 12 .. 100
```

Logiken funkar ju likadant för loopar, bara att vi får någon form av upprepning såklart.<br>
Skriv ett program som motsvarar följande flöde. Tips, använd **while** och inte **if** nu

```python
from IPython.display import Image
Image(filename="while.jpg", width = 400)
```

```python
spam = 0
while spam < 5: #När spam har räknats upp till 5 möts inte längre villkoret
    print("Hello world")
    spam = spam + 1
```
